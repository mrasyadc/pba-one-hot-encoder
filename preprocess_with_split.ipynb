```python
import json
import csv
from nltk.tokenize import sent_tokenize
```

```python
# Opening JSON file
f = open('dataset.json')
# returns JSON object as
# a dictionary
data = json.load(f)
# Iterating through the json
# list
f.close()
```

```python
count = 0

rows = []
label_list = ['Pasien', 'Usia pasien', 'Penyakit', 'Gejala', 'Kapan', 'Periode', 'Tindakan', 'Outcome', 'Pertanyaan', 'Pembuka', 'Penyebab', 'Prakondisi', 'Objek', 'Penutup', 'Referensi', 'Artikel', 'Pengukuran', 'Predikat']
headers = ['id', 'text', 'annotator', 'annotation_id', 'created_at', 'updated_at', 'lead_time'] + label_list
# print(headers)
for task in data:
    text = task['text']
    annotator = task['annotator']
    annotation_id = task['annotation_id']
    created_at = task['created_at']
    updated_at = task['updated_at']
    lead_time = task['lead_time']
    id = str(annotation_id) + '-' + str(annotator)

    print("kalimat lengkap :")
    print(text)

    sentences = sent_tokenize(text)





    if 'label' in task:
        start_pointer = 0
        end_pointer = 0
        print('all label:')
        print(task['label'])
        # untuk setiap kalimat dalam task
        for each_sentence in sentences:
            label_inside_sentence_for_each_task = set()
            start_pointer = end_pointer
            end_pointer = start_pointer + len(each_sentence) + 1
            print('start_pointer:')
            print(start_pointer)
            print('end_pointer:')
            print(end_pointer)
            print("kalimat :" + each_sentence)
            # untuk setiap label
            for labelled_text in task['label']:
                # print('labelled_text:')
                # print(labelled_text)
                label = labelled_text['labels'][0]
                start_label_index = labelled_text['start']
                end_label_index = labelled_text['end']

                if (start_label_index >= start_pointer) and (end_label_index <= end_pointer):
                    label_inside_sentence_for_each_task.add(label)
            print("labelnya:")
            print(label_inside_sentence_for_each_task)
            print("\n")

            out_list = [id+'-'+str(start_pointer), each_sentence, annotator, annotation_id, created_at, updated_at, lead_time]

            one_hot_encoding = []
            for lbl in label_list:
                if lbl in label_inside_sentence_for_each_task:
                    one_hot_encoding.append(1)
                else:
                    one_hot_encoding.append(0)
            out_list = out_list + one_hot_encoding
            rows.append(out_list)

    else:
        start_pointer = 0
        end_pointer = 0
        for each_sentence in sentences:
            start_pointer = end_pointer
            end_pointer = start_pointer + len(each_sentence) + 1
            out_list = [id+'-'+str(start_pointer), each_sentence, annotator, annotation_id, created_at, updated_at, lead_time]
            one_hot_encoding = [0] * len(label_list)
            out_list = out_list + one_hot_encoding
            rows.append(out_list)
    count = count + 1
    # if count == 100: break

#TODO: bikin output ke csv sama exceptional flow ketika label ngga ada
```

```python
with open('dataset_with_split.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(headers)

    # write multiple rows
    writer.writerows(rows)
```
